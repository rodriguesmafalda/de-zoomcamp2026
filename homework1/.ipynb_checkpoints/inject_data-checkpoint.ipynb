```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Set up connection to PostgreSQL
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
engine.connect()
```

```python
# Read parquet
import pyarrow.parquet as pq
table = pq.read_table('green_tripdata_2025-11.parquet')
print(table.schema)
```

```python
df = table.to_pandas()

df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime).dt.tz_localize(None)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime).dt.tz_localize(None)
df.store_and_fwd_flag = df.store_and_fwd_flag.astype(str)
```

```python
len(df)
```

```python
df.shape
```

```python
df.head(n=0).to_sql(name='green_trip_data', con=engine, if_exists='replace')
df.to_sql(name='green_trip_data', con=engine, if_exists='append', chunksize=10000)
```

```python
df_taxi = pd.read_csv('taxi_zone_lookup.csv')

df_taxi.head(n=0).to_sql(name='taxi_zone', con=engine, if_exists='replace')
df_taxi.to_sql(name='taxi_zone', con=engine, if_exists='append')
```
